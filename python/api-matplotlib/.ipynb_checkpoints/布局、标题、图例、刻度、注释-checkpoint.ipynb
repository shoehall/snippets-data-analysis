- 简单多图排版
- 文字注释

```python
manual_sim_repo = [[0.8333333333333334, 0.23062381852551986, 0.3333333333333333, False],
 [0.75, 0.09345794392523364, 0.75, False],
 [0.7894736842105263, 0.08403361344537816, 0.75, False],
 [0.8235294117647058, 0.6818181818181818, 0.6666666666666666, True],
 [0.8421052631578947, 0.16206261510128914, 0.6, True],
 [1.0, 0.22413793103448276, 0.8, True],
 [0.8421052631578947, 0.1748148148148148, 0.8, True],
 [0.9565217391304348, 0.9545454545454546, 1.0, True],
 [0.9090909090909091, 0.07142857142857142, 0.4375, True],
 [1.0, 0.8088888888888889, 1.0, True],
 [0.8947368421052632, 0.1261682242990654, 0.6, False],
 [0.75, 0.05984251968503937, 0.6, False]]
```

```python
%matplotlib inline

import matplotlib 
import matplotlib.pyplot as plt
import numpy as np


c = [each[3] for each in manual_sim_repo]
plt.figure(figsize=(18,5))

plt.subplot(1, 3, 1)
x = [each[0] for each in manual_sim_repo]
y = [each[2] for each in manual_sim_repo]

plt.scatter(x, y, c=c)
plt.xlabel("tags jaccard")
plt.ylabel("styles jaccard")
for i in range(len(x)):
    plt.annotate(i, xy = (x[i], y[i]), xytext = (x[i] + 0.01, y[i] + 0.01))
    
xx = np.linspace(np.min(x), np.max(x),100)
# 0.1 * yy + 0.9 * xx  = 0.8
yy = (0.8 - 0.9 * xx) / 0.1
plt.plot(xx, yy, '--')
plt.subplot(1, 3, 2)

x = [each[1] for each in manual_sim_repo]
y = [each[2] for each in manual_sim_repo]
plt.scatter(x, y, c=c)
plt.xlabel("tags seq")
plt.ylabel("styles jaccard")
for i in range(len(x)):
    plt.annotate(i, xy = (x[i], y[i]), xytext = (x[i] + 0.01, y[i] + 0.01))

plt.subplot(1, 3, 3)
x = [each[0] for each in manual_sim_repo]
y = [each[1] for each in manual_sim_repo]
plt.xlabel("tags jaccard")
plt.ylabel("tags seq")
for i in range(len(x)):
    plt.annotate(i, xy = (x[i], y[i]), xytext = (x[i] + 0.01, y[i] + 0.01))

plt.scatter(x, y, c=c)
```
